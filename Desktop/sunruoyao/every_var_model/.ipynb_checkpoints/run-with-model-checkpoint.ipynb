```python
from pyscipopt import Model, Eventhdlr, SCIP_EVENTTYPE,SCIP_HEURTIMING
import pyscipopt
import sys , os
import time
import random
import pandas as pd
import numpy as np
import joblib
import lightgbm as lgb
```

```python
print(lgb.__version__)
```

```python
class FixedVarsAtNode(Eventhdlr):
    """PySCIPOpt Event handler to write fixed vars of each node to a text file."""

    def __init__(self):

        self.count=0
        self.count_limit=-1

        self.MLmodel_path = "model/lgb_model_test1.pkl"
        
        self.MLmodel = joblib.load(self.MLmodel_path)
        print(type(self.MLmodel))

        self.nextnode = False
        

    def eventexec(self, event):
        if self.count>self.count_limit:
            self.transvars = self.model.getVars(transformed=True)
            self.depth = self.model.getDepth()
                        # start_time=time.time()
            test_info_list = []  # 用于保存收集的变量信息
            
            for var in self.transvars:
                Glb = var.getUbGlobal()  # 获取变量的全局上界
                Gub = var.getLbGlobal()  # 获取变量的全局下界
                lb = var.getLbLocal()    # 获取变量的局部下界
                ub = var.getUbLocal()    # 获取变量的局部上界
                var_type = 0 if var.vtype() == 'BINARY' else 1
            
                test_info_add_dict = {
                    'Depth': self.depth,
                    'Type': int(var_type),
                    'LowerBound': lb,
                    'UpperBound': ub,
                    'GLB': Glb,
                    'GUB': Gub
                }
                test_info_list.append(test_info_add_dict)
            
            # 使用 pd.DataFrame.from_dict 将字典列表转换为 DataFrame
            self.df_test_info = pd.DataFrame.from_dict(test_info_list)
            
            y_pred_prob = self.MLmodel.predict(self.df_test_info, num_iteration=30)
            y_pred = y_pred_prob.mean()
            # y_pred = [1 if p > 0.5 else 0 for p in y_pred_prob]
            # y_pred = [int(x) for x in y_pred]



            # y_pred_prob = self.MLmodel.predict(self.df_test_info,num_iteration=30)
            # end_time=time.time()
            # print(f"time is{start_time-end_time}")
            # y_pred = [1 if p>0.5 else 0 for p in y_pred_prob]
            # y_pred = [int(x) for x in y_pred]
            
            # 检查 y_pred 中是否有至少一个1
            if y_pred > 0.5:
                self.nextnode = True
                self.model.setIntParam("heuristics/rins/freq",1)
            else:
                self.nextnode = False
                self.model.setIntParam("heuristics/rins/freq",100)  

            self.count += 1
        else:
            pass  

    def eventinit(self):
        self.model.catchEvent(SCIP_EVENTTYPE.NODESOLVED, self)

    def eventexit(self):
        self.model.dropEvent(SCIP_EVENTTYPE.NODESOLVED, self)
```

```python
if __name__ == "__main__":
    
    test = pyscipopt.scip.Model()
    test.readProblem("/Users/oukeikou/Desktop/sunruoyao/DATA/easy-sample/markshare_4_0.mps")
    print("read done")

    # Create and add event handler with the specified output file
    eventhdlr = FixedVarsAtNode()
    test.includeEventhdlr(eventhdlr, "FixedVarsAtNode", "Python event handler to write fixed variables after each solved node")

    # Optimize the problem
    start_time = time.time()
    
    test.optimize()
    # start measuring time
    end_time = time.time()

    # calculate the processing time
    elapsed_time = end_time - start_time
    
    with open("time_marks.txt", "w") as f:
            f.write(f"with ML: {elapsed_time}\n")

    
```
