```python
for school in schoolname:
    df_tmp=df.groupby("受験校").get_group(school)
    
    pass_number_mean=df_tmp[df_tmp["合否"]==1].groupby(["年度"]).count().mean()["合否"]
    pass_number_std=df_tmp[df_tmp["合否"]==1].groupby(["年度"]).count().std()["合否"]
    pass_number_max=df_tmp[df_tmp["合否"]==1].groupby(["年度"]).count().max()["合否"]
    pass_number_min=df_tmp[df_tmp["合否"]==1].groupby(["年度"]).count().min()["合否"]
    pass_mean=df_tmp[df_tmp["合否"]==1].groupby(["年度"]).count().min()["合否"]
    
    all_number_mean=df_tmp.groupby(["年度"]).count().mean()["合否"]
    all_number_std=df_tmp.groupby(["年度"]).count().std()["合否"]
    all_number_max=df_tmp.groupby(["年度"]).count().max()["合否"]
    all_number_min=df_tmp.groupby(["年度"]).count().min()["合否"]
    dict_tmp={
        "合格平均人数":pass_number_mean,
        "合格人数分散":pass_number_std,
        "合格人数最大値":pass_number_max,
        "合格人数最小値":pass_number_min,
        "受験人数平均値":all_number_mean,
        "受験人数分散":all_number_std,
        "受験人数最大値":all_number_max,
        "受験人数最小値":all_number_min
    }
    
    for feature in subject_features:    
        dict_tmp[feature+"_mean"]=df_tmp[df_tmp["合否"]==1].groupby(["年度"])[feature].mean().mean()
        dict_tmp[feature+"_std"]=df_tmp[df_tmp["合否"]==1].groupby(["年度"])[feature].mean().std()
        dict_tmp[feature+"_min"]=df_tmp[df_tmp["合否"]==1].groupby(["年度"])[feature].mean().min()
        dict_tmp[feature+"_max"]=df_tmp[df_tmp["合否"]==1].groupby(["年度"])[feature].mean().max()
    
        dict_tmp[feature+"_min_mean"]=df_tmp[df_tmp["合否"]==1].groupby(["年度"])[feature].min().mean()
        dict_tmp[feature+"_min_std"]=df_tmp[df_tmp["合否"]==1].groupby(["年度"])[feature].min().std()
        dict_tmp[feature+"_min_min"]=df_tmp[df_tmp["合否"]==1].groupby(["年度"])[feature].min().min()
        dict_tmp[feature+"_min_max"]=df_tmp[df_tmp["合否"]==1].groupby(["年度"])[feature].min().max()


    feature_maker[school]=dict_tmp
```

```python
df_list=[]
for school_id in tqdm(range(len(schoolname))):
    df_original=df_train.groupby("受験校").get_group(schoolname[school_id])[paras]
    df_original["fake"]=0
#-------------------------------------------------------测试集/训练集区分----------------------------------------------------------------------------------
#     if len(df_test[df_test["受験校"]==schoolname[school_id]])>0:
#         test=df_test[df_test["受験校"]==schoolname[school_id]][paras]
#     else:
#         test=df_original
#为了克服数据不均衡问题，采用上采样策略
#------------------------------------------------------上采样--------------------------------------------------------------------------------   
   
    #如果合格人数为0，则判定为报名人中的最高分+20为合格的人的最高分数，否则，则以最高分数的人的分数为标准。
    num_1=len(df_original[df_original["合否"]==1])
    num_0=len(df_original[df_original["合否"]==0])
    if num_1==0:
        p_max=df_original[df_original["合否"]==0]["max(全部)"].max()+20
        p_min=df_original[df_original["合否"]==0]["max(全部)"].min() 
    else:
        p_max=df_original[df_original["合否"]==1]["max(全部)"].max()
        p_min=df_original[df_original["合否"]==1]["max(全部)"].min()
    
    #总分低于【不合格人】的最低分的人判断为不合格
    #总分低高于【合格人】的最高分的人判断为合格
    
    #从整体数据集中对符合上述要求的学生进行上采样
    df_filter_0=df_all[df_all["max(全部)"]<p_min]
    df_filter_1=df_all[df_all["max(全部)"]>p_max]
    try:
        if num_0<100:
            df_filter_0=df_filter_0[para].sample(n=100-num_0)
            df_filter_0["受験校"]=schoolname[school_id]
            df_filter_0["合否"]=0
            df_filter_0["fake"]=1
        else:
            df_tmp=df_original[df_original["合否"]==0]
            df_filter_0=df_tmp[para].sample(n=num_1)
            df_filter_0["受験校"]=schoolname[school_id]
            df_filter_0["合否"]=0
            df_filter_0["fake"]=1
    except:
        df_filter_0=df_original[df_original["合否"]==0]
    try:
        df_filter_1=df_filter_1[paras].sample(n=100-num_1)
    except:
        df_filter_1=df_filter_1[paras]

    df_filter_1["受験校"]=schoolname[school_id]
    df_filter_1["合否"]=1
    df_filter_1["fake"]=1
    result=df_filter_1.append(df_filter_0)
    result=result.append(df_original)
    result.reset_index(drop=True, inplace=True)
    df_list.append(result)
```

```python
df_train_over=pd.concat(df_list)
```
